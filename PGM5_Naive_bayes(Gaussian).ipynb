```python
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
from sklearn.model_selection import train_test_split

data = pd.read_csv("P5_naivedata.csv")

print(data.head())

features = list(data.columns)[:-1]
predicted = list(data.columns)[-1]

X = data[features].values
Y = data[predicted].values

xtrain,xtest,ytrain,ytest = train_test_split(X,Y)

print("Training examples: ",xtrain.shape[0])
print("Testing examples: ",xtest.shape[0])

clf = GaussianNB().fit(xtrain,ytrain)
p = clf.predict(xtest)
for i in range(len(p)):
    print(p[i]," ",ytest[i])

print("Accuracy: ",(metrics.accuracy_score(p,ytest)*100),"%")
```
